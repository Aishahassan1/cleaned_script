```python
import os
import numpy as np
import pandas as pd
import petl as etl
```

```python
def extract_branch_name(filename):
    branch_name_split = filename.split('.')
    # removing file extension
    branch_name = branch_name_split[0]
    # splitting by underscore to remove date range
    branch_name_split2 = branch_name.split('_')
    # extracting branch name
    branch_name_split = branch_name_split2[1] + ' ' + branch_name_split2[2]
    return branch_name_split
```

```python
# all the code below is looping through all the store records with the code added to this section

directory_contents = os.listdir()
# read product list
products_list_table = etl.fromcsv('additional_files/products_list.csv')
# read branch list
branch_list_table = etl.fromcsv('additional_files/branch_list.csv')
for file in directory_contents:
    if file.endswith('.csv') :
        print(file) 
        branch_name = extract_branch_name(file)
        # read store record
        store_record_table = etl.fromcsv(file)
        # update file headers - remove inconsistencies in header naming
        store_record_table = etl.setheader(store_record_table, ['year', 'month', 'day', 'hour', 'product', 'quantity', 'amount'])
        # adding branch name to data
        store_record_table = etl.addfield(store_record_table, 'branch_name', branch_name)
        # clean data - remove non numeric data
        store_record_table = etl.select(store_record_table , 'quantity', lambda quantity: str(quantity).isnumeric())
        store_record_table = etl.select(store_record_table , 'quantity', lambda quantity: int(quantity) != 0)
        store_record_table = etl.select(store_record_table , 'amount', lambda amount: str(amount).replace('.','').isnumeric())
#         merge store record and product list record to include product category
        store_record_table = etl.join(store_record_table, products_list_table, key = 'product')
        # merge store record and branch list record to include branch name 
        store_record_table = etl.join(store_record_table, branch_list_table, key = 'branch_name')
        # save cleaned store record containing product category
#         store_record_table = etljoin(store_record_table, branch_expenses_table, key = 'branch_name')
        etl.tocsv(store_record_table,'cleaned/'+file) 
        store_record_table = None
        print('cleaned/'+file)
        #clean file here
        print('completed1')
```
